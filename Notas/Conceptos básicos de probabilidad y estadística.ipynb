## Variables aleatorias

Las variables aleatorias $X$ son funciones reales que asignan un valor, (lógico, numérico, función), al resultado de un experimento aleatorio.

$$X:\Omega\rightarrow\mathbb{R},$$

donde $\Omega$ es el espacio muestral.

Los valores posibles de una variable aleatoria pueden representar los posibles resultados de un experimento aún no realizado, o los posibles valores de una cantidad cuyo valor actualmente existente es incierto (p.e., como resultado de una medición incompleta o imprecisa). Intuitivamente, una variable aleatoria puede tomarse como una cantidad cuyo valor no es fijo pero puede tomar diferentes valores; una distribución de probabilidad se usa para describir la probabilidad de que se den los diferentes valores. En términos formales una variable aleatoria es una función definida sobre un espacio de probabilidad.

### Tipos de variables
#### Variable aleatoria discreta
Es discreta si su recorrido es un conjunto discreto.

#### Variable aleatoria continua
Es continua si su recorrido es un conjunto no numerable.

Para variables aleatorias discretas, la medida de probabilidad se define a través de la función de masa

$$p(x)=P[X=a]$$

#### Propiedades

$$p(x) \geq 0$$
$$\sum p(x_i) = 1$$

### Función de distribución

Se define como

$$F(a) = \sum p(x) = P[X \leq a]$$

#### Propiedades
<ol>
    <li> Es continua por la derecha
        </li>
    <li> Es monótona no decreciente
        </li>
    <li> $\lim_{x\rightarrow -\infty} = 0$ y $\lim_{x\rightarrow \infty} = 1$
        </li>
    </ol>
    
### Esperanza
La esperanza o valor esperado de una variable aleatoria es la suma del producto de la probabilidad de cada suceso por el valor de dicho suceso. Si todos los sucesos son de igual probabilidad entonces la esperanza es la media aritmética.

#### Caso discreto
$$\mu = E[X] = \sum x_ip(x_i)$$

#### Caso continuo
$$\mu = E[x] = \int_\Omega XdP$$

Si la v.a. es una función $Y=g(X)$, entonces

$$\mu = E[Y] = \sum g(x_i)p(x_i)$$

### Propiedades
<ol>
    <li> Es lineal.
    </li>
    <li> $E[X^n] = \sum x_i^n p(x_i)$ (centrado en cero$\rightarrow$ medida de tendencia central)
    </li>
</ol>

### Varianza

Intuitivamente, se refiere al centro de gravedad de la distribución. Hace referencia a la dispersión en la distribución.

$$E[|X-E[x]] \rightarrow Var(X) = E[(X-E[X])^2] = E[X^2]-E^2[X]$$

### Desviación estándar
$$DE[X] = \pm \sqrt Var[X]$$

### Covarianza
Es un valor que indica el grado (tendencia) de variación conjunta de dos variables aleatorias $X$ e $Y$ respecto a sus valores medios. Se expresa como

$$Cov(X,Y) = E[X-E[X]]E[Y-E[Y]]=E[XY]-E[X]E[Y]$$

Si $X$ e $Y$ son v.a. independientes, entonces $E[XY]=E[X]E[Y]$.

Si $X$ e $Y$ son v.a. independientes, $Cov(X,Y)=0$. Lo recíproco no implica independencia.


## Medidas de tendencia central y dispersión

### Promedio
Es un valor típico o representativo de un conjunto de datos. Como estos valores típicos tienden a encontrarse en el centro de los conjuntos de datos, ordenados de acuerdo con su magnitud, a los promedios se les conoce también como medidas de tendencia central.

Se pueden definir varios tipos de promedios; los más usados son la media aritmética, la mediana, la moda, la media
geométrica y la media armónica.

#### Media aritmética
La media aritmética, o brevemente la media, de un conjunto de $N$ números $X_1$, $X_2$, $X_3$, $\cdots$ , $X_N$ se denota así: $\bar{X}$ y está definida como

$$ \bar{X} = \frac{\sum_i^N X_i}{N}$$

#### Media aritmética ponderada
Algunas veces, se asignan ciertos factores de ponderación (o pesos) $w_i$ que dependen del significado o importancia que se les asigne a estos números. En este caso, la media aritmética ponderada es

$$ \bar{X} = \frac{\sum_i^N w_i X_i}{\sum w}$$

#### Desviación respecto a la media
Es la diferencia en valor absoluto entre cada valor de la variable estadística y la media aritmética.

$$\Delta_i = |X_i - \bar{X}|$$

#### Varianza
Es la media aritmética del cuadrado de las desviaciones respecto a la media de una distribución estadística. La varianza intenta describir la dispersión de los datos.

$$ \sigma^2 = \frac{\sum_i \Delta_i^2}{N}$$

#### Desviación típica
Es la raíz cuadrada de la varianza.

En las distribuciones normales se encuentra que:
<ol>
<li> 68.27% de los casos está comprendido entre $\bar{X}−\sigma$ y $\bar{X}+\sigma$ (es decir, una desviación estándar a cada lado de
la media).</li>
<li> 95.45% de los casos está comprendido entre $\bar{X}−2\sigma$ y $\bar{X}+2\sigma$ (es decir, dos desviaciones estándar a cada lado de la media). </li>
<li> 99.73% de los casos está comprendido entre $\bar{X}−3\sigma$ y $\bar{X}+3\sigma$ (es decir, tres desviaciones estándar a cada lado
de la media). </li>
</ol>
 En distribuciones moderadamente sesgadas, estos porcentajes se satisfacen de manera aproximada.

#### Mediana
La mediana de un conjunto de números acomodados en orden de magnitud (es decir, en una ordenación) es el valor
central o la media de los dos valores centrales.

#### Moda
La moda de un conjunto de números es el valor que se presenta con más frecuencia; es decir, es el valor más frecuente. Puede no haber moda y cuando la hay, puede no ser única.

### Relación empírica entre media, mediana y moda
En las curvas de frecuencias unimodales que son ligeramente sesgadas (asimétricas), se tiene la relación empírica
siguiente:
$$ Media - moda = 3(media - mediana)$$




#### Sesgo

El sesgo de una distribución es su grado de asimetría o el grado en el que se aleja de la simetría. Si una curva de frecuencias de una distribución tiene una cola más larga hacia la derecha del máximo central que hacia la izquierda, se dice que la distribución es sesgada a la derecha, o que tiene un sesgo positivo. Si ocurre lo contrario, se dice que es sesgada a la izquierda o que tiene un sesgo negativo.
En las distribuciones sesgadas, la media tiende a encontrarse del mismo lado que la cola más larga opuesto al de la moda y que la cola más larga. Por lo tanto, una medida de la simetría (o sesgo) se obtiene mediante la diferencia: media – moda. Esta medida se puede hacer adimensional dividiendo entre una medida de dispersión, como la desviación estándar, lo que conduce a la definición (primer coeficiente de sesgo de Pearson):
$$sesgo = \frac{\bar{X}-moda}{\sigma}$$

## Teoría estadística de la decisión

En la práctica, con frecuencia se tienen que tomar decisiones acerca de una población con base en información muestral. A tales decisiones se les llama decisiones estadísticas.

Cuando se trata de tomar una decisión es útil hacer suposiciones (o conjeturas) acerca de la población de que se trata.
A estas suposiciones, que pueden ser o no ciertas, se les llama hipótesis estadísticas. Estas hipótesis estadísticas son por lo general afirmaciones acerca de las distribuciones de probabilidad de las poblaciones.

### Hipótesis nula
En muchas ocasiones se formula una hipótesis estadística con la única finalidad de refutarla o anularla. Por ejemplo, si se quiere decidir si una moneda está cargada o no, se formula la hipótesis de que no está cargada (es decir, p = 0.05, donde p es la probabilidad de cara). También, si se quiere decidir si un método es mejor que otro, se formula la hipótesis de que no hay diferencia entre los dos (es decir, que cualquier diferencia que se observe se debe sólo a las fluctuaciones del muestreo de una misma población). A estas hipótesis se les llama hipótesis nula y se denota $H_0$.

### Regla de decisión
Si se supone que una hipótesis es verdadera, pero se encuentra que los resultados que se observan en una muestra
aleatoria difieren marcadamente de los resultados esperados de acuerdo con la hipótesis, entonces se dice que las diferencias observadas son significativas y se estará inclinado a rechazar la hipótesis (o por lo menos a no aceptarla de acuerdo con la evidencia obtenida). Por ejemplo, si en 20 lanzamientos de una moneda se obtienen 16 caras, se estará inclinado a rechazar que la moneda es buena, aun cuando se puede estar equivocado.
A los procedimientos que permiten determinar si las muestras observadas difieren significativamente de los resultados esperados, ayudando así a decidir si se acepta o se rechaza la hipótesis, se les llama pruebas de hipótesis, pruebas de significancia o reglas de decisión.

### Nivel de significancia
Se acostumbran los niveles de significancia 0.05 o 0.01, aunque también se usan otros valores.

### Usando distribución normal

Considerando el estadístico $S$, la media $\mu$ y la desviación estándar $\sigma$ de una distribución, la distribución de la variable estandarizada (o puntuación $z$), está dada por

$$z = \frac{S-\mu}{\sigma}$$

Se puede tener una confianza del 95% en que si la hipótesis es verdadera, entonces la puntuación $z$ del estadístico muestral $S$ estará entre −1.96 y 1.96. Pero si se toma una sola muestra aleatoria y se encuentra que la puntuación $z$ del estadístico se encuentra fuera del rango −1.96 a 1.96, se concluye que si la hipótesis dada es verdadera, esto sólo puede ocurrir con una probabilidad de 0.05.



### Prueba de una y dos colas

Si es de interés los valores extremos del estadístico $S$, o de sus correspondientes puntuaciones $z$, a ambos lados de la media (es decir, en las dos colas de la distribución), entonnces a estas pruebas se les llama pruebas bilaterales o pruebas de dos colas.
Sin embargo, hay ocasiones en las que interesan únicamente los valores extremos a un solo lado de la media (es
decir, en una sola cola de la distribución); por ejemplo, cuando se prueba si un método es mejor que otro (que es distinto a probar si un método es mejor o peor que otro). A este tipo de pruebas se les llama pruebas unilaterales o pruebas de una cola. En estos casos la región crítica es una región en un solo lado de la distribución y su área es igual al nivel de significancia.

### Valor p
El valor $p$ es la probabilidad de obtener un estadístico muestral tan extremo o más extremo que el obtenido, suponiendo que la hipótesis nula sea verdadera. Para probar una hipótesis empleando este método se establece un valor $\alpha$, si $p \leq \alpha$ se rechaza la hipótesis nula.

Sean $\bar{X_1}$ y $\bar{X_2}$ las medias muestrales de muestras grandes de tamaños $N_1$ y $N_2$ obtenidas de poblaciones cuyas medias son $\mu_1$ y $\mu_2$ y cuyas desviaciones estándar son $\sigma_1$ y $\sigma_2$, respectivamente. Considérese la hipótesis nula de que no hay diferencia entre las dos medias poblacionales (es decir, $\mu_1$ = $\mu_2$), lo cual es equivalente a decir que las muestras se han tomado de dos poblaciones que tienen la misma media.

#### Caso: Diferencia entre medias
$$\mu_{X_1 - X_2} = 0,$$
$$\sigma_{X_1 - X_2} = \sqrt(\frac{\sigma_1^2}{N_1}+\frac{\sigma_2^2}{N_2}),$$
entonces
$$z = \frac{\bar{X_1}-\bar{X_2}}{\sigma_{X_1 - X_2}}$$

## Correlación
Establece la relación o dependencia que existe entre las dos variables que intervienen en una distribución bidimensional. Determina si los cambios en una de las variables influyen en los cambios de la otra. En caso de que suceda, diremos que las variables están correlacionadas o que hay correlación entre ellas. La correlación es positiva cuando los valores de las variables aumenta juntos; y es negativa cuando un valor de una variable se reduce cuando el valor de la otra variable aumenta.

**Componentes**

<ol>
    <li> La fuerza mide el grado en que la línea representa a la nube de puntos: si la nube es estrecha y alargada, se representa por una línea recta, lo que indica que la relación es fuerte; si la nube de puntos tiene una tendencia elíptica o circular, la relación es débil.
    </li>
    <li> El sentido mide la variación de los valores de B con respecto a A: si al crecer los valores de A lo hacen los de B, la relación es directa (pendiente positiva); si al crecer los valores de A disminuyen los de B, la relación es inversa (pendiente negativa).
    </li>
</ol>

### Coeficiente de correlación de Pearson

$$r = \frac{\sum_i^N (\Delta X_i)(\Delta Y_i)}{\sqrt(\sum_i^N \Delta X_i)\sqrt(\sum_i^N \Delta Y_i)}$$

Bajo el supuesto de que $X$ e $Y$ se extraen de distribuciones normales independientes (por lo que el coeficiente de correlación de la población es 0), y considerando $n$ como el número de muestras, $B$ la función beta, entonces la función de densidad de probabilidad del coeficiente de correlación de la muestra $r$ es

$$f(r) = \frac{(1-r^2)^{(n/2)-2}}{B(\frac{1}{2}, \frac{n}{2}-1)}$$

### Coeficiente de correlación de Spearman
Es una medida de la correlación entre dos variables aleatorias (tanto continuas como discretas). Para calcular $\rho$, los datos son ordenados y reemplazados por su respectivo orden. Considerando $D$ como la diferencia entre los estadísticos $X$ y $Y$, y $N$ el número de parejas de datos:

$$ \rho = 1 - \frac{6\sum D^2}{N(N^2-1)}$$
