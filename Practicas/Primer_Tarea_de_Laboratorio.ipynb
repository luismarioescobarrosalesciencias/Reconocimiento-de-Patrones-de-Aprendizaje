3.Use el método Newton-Raphson para encontrar una raíz$(f (x) = 0)$ y especifiqué el valor inicial $x_0 =?$ usado, también
especifiqué la derivada de:(Nota no usar el valor de la raíz en $x_0$)

a) Ejemplo $f (x) = sen(x)$ con valor inicial $x_0 = 3$ y derivada $f ′(x) = cos(x)$ da :

Aproximación de la solución de $sin(x) = 0$:
3.141592653589793.

b) $f (x) = cos(x)$

c) $f (x) = xsen(2x) + 1$

d) $f (x) = x^2 + 5x + 6$

e) $f (x) = x + e^x$

```python
#Codigo

def newthon_raphson(f:Callable,df:Callable,x0:int,tol:float,max_iter:int):
  x = x0
  for i in range(max_iter):
    fx = f(x)
    if abs(fx) < tol:
      return x
    dfx = df(x)
    if dfx == 0 :
      raise Exception("No se puede calcula la raiz")
    x = x - (fx/dfx)
  return None


''''''''''''''''a) f(x) = sen(x)'''''''''''''''''''''''''''''
x0 = 3  # Valor inicial
tolerancia = 1e-8  # Tolerancia
max_iteraciones = 100  # Número máximo de iteraciones
'''
f(x )= sen(x)
f'(x) = cos(x)
'''
raiz_ = newthon_raphson(lambda x: np.sin(x), lambda x: np.cos(x), x0, tolerancia, max_iteraciones)
print("a) Aproximación de la solución de sin(x): ",raiz_)

''''''''''''''''b) f(x) = cos(x)'''''''''''''''''''''''''''''
x0 = 7 # Valor inicial
tolerancia = 1e-8  # Tolerancia
max_iteraciones = 100  # Número máximo de iteraciones
'''
f(x )= cos(x)
f'(x) = -sen(x)
'''
raiz_ = newthon_raphson(lambda x: np.cos(x), lambda x: -np.sin(x), x0, tolerancia, max_iteraciones)
print("b) Aproximación de la solución de cos(x): ",raiz_)

''''''''''''''''c) f(x) = xsen(2x) + 1'''''''''''''''''''''''''''''
x0 = 5 # Valor inicial
tolerancia = 1e-8  # Tolerancia
max_iteraciones = 100  # Número máximo de iteraciones
'''
f(x )= xsen(2x) + 1
f'(x) = 2xcos(2x) + sin(2x)
'''
raiz_ = newthon_raphson(lambda x: (x* np.sin(2*x)) + 1, lambda x: (2*x*np.cos(2*x))+np.sin(2*x), x0, tolerancia, max_iteraciones)
print("c) Aproximación de la solución de xsen(2x)+1: ",raiz_)

''''''''''''''''a) f(x) = x^2+5x+6'''''''''''''''''''''''''''''
x0 = 4 # Valor inicial
tolerancia = 1e-8  # Tolerancia
max_iteraciones = 100  # Número máximo de iteraciones
'''
f(x )= x^2+5x+6
f'(x) = 2x+5
'''
raiz_ = newthon_raphson(lambda x: (x**2)+(5*x)+6, lambda x: (2*x)+5, x0, tolerancia, max_iteraciones)
print("d) Aproximación de la solución de x^2+5x+6: ",raiz_)

''''''''''''''''a) f(x) = x + e^x  '''''''''''''''''''''''''''''
x0 = 2# Valor inicial
tolerancia = 1e-8  # Tolerancia
max_iteraciones = 100  # Número máximo de iteraciones
'''
f(x )= x + e^x
f'(x) = 1 + e^x
'''
raiz_ = newthon_raphson(lambda x: np.exp(x) + x, lambda x: (1 + np.exp(x)),x0, tolerancia, max_iteraciones)
print("d) Aproximación de la solución de x + e^x: ",raiz_)




```

4.Usando python y la ecuación de la recta.Gráfica la recta de puntos que pasan por además indica la pendiente de la recta:

a) (0, 2), (1, −2)

b) (1, 2), (−1, −4)

c) (2, 2), (1, −6)

d) (3, 2), (−1, −8)

```python
#Codigo

#Metodo para calcular la pendiente

def pendiente(punto1: tuple[float, float], punto2: tuple[float, float]) -> float:
  if punto1[0] == punto2[0] and punto1[1] == punto2[1]:
    raise Exception("Los puntos son la misma coordenada ")

  return (punto2[1]-punto1[1])/(punto2[0]-punto1[0])

def ecuacion_recta(pendiente:float, punto:tuple[float, float])-> Callable:
  return lambda x: pendiente* (x-punto[0])+punto[1]



def grafica_recta_ecuacion(punto1: tuple[float, float], punto2: tuple[float, float]):
  x = np.array([punto1[0],punto2[0]])
  y = np.array([punto1[1],punto2[1]])

  m = pendiente(punto1,punto2)

  x_min = min(x)
  x_max = max(x)

  # Expandimos el rango de impresion de la grafica
  x_new = np.linspace(x_min - 1, x_max + 1, 10)
  y_new = ecuacion_recta(pendiente(punto1, punto2), (x[0], y[0]))(x_new)
  plt.plot(x_new,y_new,label=f'Pendiente = {m:.2f}')



#Metodo que grafica las rectas usando matplot , para comprobar que los metodos anteriores son correctos
def graficar_recta(punto1: tuple[float, float], punto2: tuple[float, float]):
  m = pendiente(punto1,punto2)

  x_values = [punto1[0], punto2[0]]
  y_values = [punto1[1], punto2[1]]

  plt.plot(x_values, y_values, label=f'Pendiente = {m:.2f}')
  plt.xlim(min(punto1[0], punto2[0]) - 1, max(punto1[0], punto2[0]) + 1)
  plt.ylim(min(punto1[1], punto2[1]) - 1, max(punto1[1], punto2[1]) + 1)

```

6.Genere una Interpolación Polinómica de Newton usando $n$ puntos conocidos de las funciones:

a) $cos(x)$ Sobre el intervalo $[0, 2π]$. Indica cuantos puntos $n$ usar para que el Error absoluto medio sea menor que
$0.01$ comparando a la función $np.cos(x)$ como Valores objetivo de verdad fundamental (correctos) y la Interpolación
Polinómica como Valores objetivo estimados.

b) $tan(x)$ Sobre el intervalo $[−π/2, π/2]$. Indica cuantos puntos $n$ usar para que el Error absoluto medio sea menor que
$0.01$ comparando a la función $np.tan(x)$ como Valores objetivo de verdad fundamental (correctos) y la Interpolación
Polinómica como Valores objetivo estimados.

```python
#Codigo

def diferencias_divididas(x, y):
  n = len(x)
  coef = np.array(y, dtype=float)
  coef
  for j in range(1, n):
    for i in range(n-1, j-1, -1):
        #calulamos las diferencias dividas
        coef[i] = (coef[i] - coef[i-1]) / (x[i] - x[i-j])
  return coef

coef_aux = diferencias_divididas(x, y)

def newton_interpolation(x, coef, x_new):
  n = len(x)
  p = coef[n-1]
  for i in range(n-2, -1, -1):
    p = coef[i] + (x_new - x[i]) * p
  return p

n_con = 6
x = np.linspace(0,2*np.pi,n_con) # Este es el n= que hay que buscar.
y = np.cos(x) # Para la tarea sustituye por cos(x), o tan(x).

x_new = np.linspace(0, 2*np.pi, 100)
coef_aux = diferencias_divididas(x, y)

x_sym = sym.Symbol('x')
polinomio_newton = newton_interpolation(x, coef_aux, x_sym)

# simplifica el polinomio
polisimple = polinomio_newton.expand()

# simplifica el polinomio para visualizar lo comodamente
polisimple_round = sym.Poly.from_list([np.format_float_scientific(coff, precision=1, exp_digits=2) for coff in polisimple.as_poly().all_coeffs()], x_sym).as_expr()


y_real = np.cos(x_new)
#y_new = newton_interpolation(x, coef_aux, x_new)
y_new = newton_interpolation(x, coef_aux, x_new)

plt.plot(x_new, y_new, label='Polinomio de Newton')
plt.plot(x, y, 'o', label='Valores conocidos (Interpolación Lineal)')
plt.plot(x_new, np.cos(x_new), '--', label='Cos')
plt.title(f'Interpolación polinómica de Newton \n N.Con: {n_con} \n $p(x) = {str(polisimple_round)[:30]}$ ... \n MAE: {mean_absolute_error(y_real,y_new)}')
plt.legend()

plt.show()

```

b) tan(x) Sobre el intervalo [−π/2, π/2]. Indica cuantos puntos n usar para que el Error absoluto medio sea menor que 0.01 comparando a la función np.tan(x) como Valores objetivo de verdad fundamental (correctos) y la Interpolación
Polinómica como Valores objetivo estimados.

```python
#Codigo
'''Coseno'''
#Lagrange para la funcion seno
xi = np.linspace(0,2*np.pi,7)
fi = np.cos(xi)

# PROCEDIMIENTO
# Polinomio de Lagrange
n = len(xi)
x = sym.Symbol('x')
polinomio = 0
divisorL = np.zeros(n, dtype = float)
for i in range(0,n,1):

    # Termino de Lagrange
    numerador = 1
    denominador = 1
    for j  in range(0,n,1):
        if (j!=i):
            numerador = numerador*(x-xi[j])
            denominador = denominador*(xi[i]-xi[j])
    terminoLi = numerador/denominador

    polinomio = polinomio + terminoLi*fi[i]
    divisorL[i] = denominador

# simplifica el polinomio
polisimple = polinomio.expand()
# para evaluación numérica
px = sym.lambdify(x,polisimple)
#Polinomio más simple.
polisimple_round = sym.Poly.from_list([np.format_float_scientific(coff, precision=1, exp_digits=2) for coff in polisimple.as_poly().all_coeffs()], x_sym).as_expr()

# Puntos para la gráfica
muestras = 100
a = np.min(xi)-1
b = np.max(xi)+1
pxi = np.linspace(a,b,muestras)
pfi = px(pxi)

# Gráfica
plt.plot(pxi,pfi, label = 'Forma de Lagrange')
plt.plot(xi,fi,'o', label = 'Valores Conocidos')
plt.plot(pxi, np.cos(pxi), '--', label='Coseno')
plt.legend()
plt.title(f'Interpolación Lagrange \n $p(x)= {str(polisimple_round)[:30]}$')
plt.show()
```

```python
def lagrange_interpolation_mae(target_error):
    n = 2  # Empezamos con un número bajo de puntos
    while True:
        xi = np.linspace(0, 2*np.pi, n)
        fi = np.cos(xi)

        x = sym.Symbol('x')
        polinomio = 0

        for i in range(n):
            numerador = 1
            denominador = 1

            for j in range(n):
                if j != i:
                    numerador *= (x - xi[j])
                    denominador *= (xi[i] - xi[j])

            terminoLi = numerador / denominador
            polinomio += terminoLi * fi[i]

        px = sym.lambdify(x, polinomio)

        muestras = 1000
        a = np.min(xi) - 1
        b = np.max(xi) + 1
        pxi = np.linspace(a, b, muestras)
        pfi = px(pxi)

        true_values = np.cos(pxi)
        absolute_errors = np.abs(pfi - true_values)
        mae = np.mean(absolute_errors)

        if mae < target_error:
            return n

        n += 1

# Encontrar n para un MAE menor que 0.01
n_con = lagrange_interpolation_mae(0.01)
print("Número de puntos necesarios para un MAE < 0.01:", n_con)
```

```python
#Codigo
''' COSENO'''
x = np.linspace(0, 2*np.pi, 10)
y = np.cos(x)

# Definir la función del spline lineal
def linear_spline(x, y, x_eval):
    y_eval = np.zeros_like(x_eval)
    for i, x_i in enumerate(x_eval):
        j = np.searchsorted(x, x_i)
        if j == 0:
            y_eval[i] = y[0]
        elif j == len(x):
            y_eval[i] = y[-1]
        else:
            x_left, x_right = x[j-1], x[j]
            y_left, y_right = y[j-1], y[j]
            m = (y_right - y_left) / (x_right - x_left)
            y_eval[i] = y_left + m * (x_i - x_left)
    return y_eval

# Definir los puntos de evaluación del spline lineal
x_eval = np.linspace(-np.pi, 2*np.pi+np.pi, 100)

# Calcular los valores del spline lineal
y_eval = linear_spline(x, y, x_eval)

# Graficar los puntos originales y el spline lineal
plt.plot(x, y, 'o', label='Puntos')
plt.plot(x_eval, y_eval, label='Spline lineal')
plt.legend()
plt.show()
```
